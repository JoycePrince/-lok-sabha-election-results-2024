```python
party_seats = df.groupby('Party')['Seats Won'].sum()

# Create a pie chart
fig1, ax1 = plt.subplots(figsize=(10, 10))
ax1.pie(party_seats, labels=party_seats.index, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title("Seats Won by Each Party in Lok Sabha Election 2024")
plt.show()
```

```python
seatsSorted = df.sort_values(by='Seats Won', ascending=False)

# Select the top 5 and bottom 5 parties
top_5_parties = seatsSorted.head(5)
bottom_5_parties = seatsSorted.tail(5)

# Plot a bar chart for the top 5 and bottom 5 parties
plt.figure(figsize=(14, 7))

# Plotting top 5 parties
plt.bar(top_5_parties['Party'], top_5_parties['Seats Won'], label='Top 5', color='g')

# Plotting bottom 5 parties
plt.bar(bottom_5_parties['Party'], bottom_5_parties['Seats Won'], label='Bottom 5', color='b')

plt.xticks(rotation=45, ha='right')

plt.xlabel("Party")
plt.ylabel("Seats Won")
plt.title("Top 5 and Bottom 5 Parties in Lok Sabha Election 2024 by Seats Won")
plt.legend()
plt.show()
```
